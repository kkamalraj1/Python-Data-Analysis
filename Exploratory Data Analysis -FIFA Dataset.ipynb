### Finding the best goal keeper

#### Characteristic of Goal Keeper
#### Shot Stopper
#### Sweeper

```python
#weights
a = 0.5
b = 1
c= 2
d = 3
 
#GoalKeeping Characterstics
df['gk_Shot_Stopper'] = (b*df.Reactions + b*df.Composure + a*df.Speed + a*df.Strength + c*df.Jumping + b*df.GK_Positioning + c*df.GK_Diving + d*df.GK_Reflexes + b*df.GK_Handling)/(2*a + 4*b + 2*c + 1*d)
df['gk_Sweeper'] = (b*df.Reactions + b*df.Composure + b*df.Speed + a*df.Short_Pass + a*df.Long_Pass + b*df.Jumping + b*df.GK_Positioning + b*df.GK_Diving + d*df.GK_Reflexes + b*df.GK_Handling + d*df.GK_Kicking + c*df.Vision)/(2*a + 4*b + 3*c + 2*d)
```

```python
plt.figure(figsize=(15,6))
df1 = df.sort_values(['gk_Shot_Stopper'],ascending=False).head(5)
x1=np.array(list(df1['Name']))
y1=np.array(list(df1['gk_Shot_Stopper']))
plt.bar(x1,y1,color='orange')
plt.xlabel('Player Name')
plt.ylabel('Shot Stopper Score')
plt.title("Top 5 Goal Keeper - Shot Stopper")
```

```python
plt.figure(figsize=(15,6))
df2 = df.sort_values(['gk_Sweeper'],ascending=False).head(5)
x2=np.array(list(df2['Name']))
y2=np.array(list(df2['gk_Sweeper']))
plt.bar(x2,y2,color='blue')
plt.xlabel('Player Name')
plt.ylabel('Sweeper')
plt.title("Top 5 Goal Keeper - Sweeper")
```

```python
img = mpimg.imread('Manuel Neuer.png')
plt.title('Best Goal Keeper')
plt.xlabel('Manuel Neuer')
plt.imshow(img)
```

### Finding the Best Defenders

##### Center Backs: We will need two center backs. One who plays Left Center Back (LCB) and the other who plays Right Center Back (RCB)
##### Wing Backs: We will need two wing backs. One who plays Left Wing Back (LWB) and the other who plays Right Wing Back (RWB)

```python
df['df_centre_backs'] = ( d*df.Reactions + c*df.Interceptions + d*df.Sliding_Tackle + d*df.Standing_Tackle + b*df.Vision+ b*df.Composure + b*df.Crossing +a*df.Short_Pass + b*df.Long_Pass+ c*df.Acceleration + b*df.Speed
+ d*df.Stamina + d*df.Jumping + d*df.Heading + b*df.Long_Shots + d*df.Marking + c*df.Aggression)/(6*b + 3*c + 7*d)
df['df_wb_Wing_Backs'] = (b*df.Ball_Control + a*df.Dribbling + a*df.Marking + d*df.Sliding_Tackle + d*df.Standing_Tackle + a*df.Attacking_Position + c*df.Vision + c*df.Crossing + b*df.Short_Pass + c*df.Long_Pass + d*df.Acceleration +d*df.Speed + c*df.Stamina + a*df.Finishing)/(4*a + 2*b + 4*c + 4*d)
```

#### Left Central Defender

```python
plt.figure(figsize=(15,6))
df3 = df[(df['Club_Position'] == 'LCB')].sort_values('df_centre_backs', ascending=False).head(5)
x3 = np.array(list(df3['Name']))
y3 = np.array(list(df3['df_centre_backs']))
plt.bar(x3, y3)
plt.xlabel("Player Name")
plt.ylabel("LCB Score")
plt.title("Left Central Defender")
```
